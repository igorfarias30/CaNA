```python
import matplotlib.pyplot as plt
from asterisco import asterisco
import numpy as np
import time
```

```python
constants = np.arange(1, 10000, 10)
```

```python
times = []
for constant in constants:
    start = time.time()
    _ = asterisco(constant)
    times.append(time.time() - start)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(times)
plt.title("Time Asterisc (sec)")
plt.xlabel("N size")
plt.ylabel("Time")
plt.show()
```
