```python
import matplotlib.pyplot as plt
from asterisco import asterisco
import numpy as np
import time
```

```python
constants = np.arange(1, 10000, 5)
```

```python
times = []
executions = []

for constant in constants:
    start = time.time()
    executions.append(asterisco(constant))
    times.append(time.time() - start)
```

```python
fig, ax1 = plt.subplots(figsize=(16, 8))

color = 'tab:red'
ax1.set_xlabel('N - size')
ax1.set_ylabel('time (s)', color=color)
ax1.plot(constants, times, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('executions', color=color)  # we already handled the x-label with ax1
ax2.plot(constants, executions, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.grid()
plt.show()
```
